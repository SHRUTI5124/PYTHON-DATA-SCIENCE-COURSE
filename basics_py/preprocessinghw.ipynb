```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
pd.set_option('display.max_columns',None)
```

```python
df = sns.load_dataset('penguins')
df
```

```python
df.isnull().sum()
```

```python
num_missing_data=['bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g']
cat_missing_data=['sex']
```

```python
for col in num_missing_data:
    df[col]=df[col].astype(float)
    df[col].fillna(df[col].mean(),inplace=True)
for col in cat_missing_data:
    df[col].fillna(df['sex'].value_counts().idxmax(),inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df.rename(columns={
    'bill_length_mm':'length(mm)',
    'bill_depth_mm':'depth(mm)',
    'flipper_length_mm':'fl(mm)',
    'body_mass_g':'mass(g)'
},inplace=True)
df
```
