```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('laptops (1).csv')
df
```

```python
def get_brand_name(name):
    brand=name.split()[0]
    brand=brand.lower()
    return brand
```

```python
get_brand_name('APPLE MacBook Air M1 - (16 GB/256 GB SSD/Mac O...	')
```

```python
df['brand']=df['name'].apply(get_brand_name)
df
```

```python
def clean_price_coloumn(price):
    if isinstance(price,str):
        price=price.replace("₹",'').replace(",",'')
        price=int(price)
    return price
```

```python
clean_price_coloumn('₹40,990')
```

```python
df['price'].apply(clean_price_coloumn)
```

```python
df['clean_price']=df['price'].apply(clean_price_coloumn)
df
```

```python
def get_num_of_rating(rr):
    if isinstance(rr,str):
        nor=rr.split()[0]
        nor=nor.replace(',','')
        nor=int(nor)
        return nor
    else:
        np.nan

def get_num_of_reveiws(rr):
    if isinstance(rr,str):
        x=rr.split('&')[-1]
        x=x.split()[0]
        x=x.replace(',','')
        x=int(x)
        return x
    else:
        np.nan
```

```python
df['rating']=df['rr'].apply(get_num_of_rating)
df['review']=df['rr'].apply(get_num_of_reveiws)
df
```
